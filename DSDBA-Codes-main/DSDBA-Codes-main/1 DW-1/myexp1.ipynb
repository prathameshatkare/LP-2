DROPPING NULL OR MISSING VALUES This is the fastest and easiest step to handle missing
values. However, it is not generally advised. This method reduces the
quality of our model as it
reduces sample size because it works by deleting all other observations where any of the variable
is missing. The process can be done by:
data_name.dropna()

```python
df.dropna()
```

It will be observed that of 10 entries will be reduced to 3 just by dropping NaN values.!!! Dropping is
only advised to be used if missing values are few (say 0.01–0.5% of our
data).

```python
#original dataset is not changed
df
```

```python
#We can drop columns that have at least one NaN in any row by setting the axis argument
df.dropna(axis=1)
```

```python
#The dropna() method has several additional parameters:
df.dropna(how='all')
```

```python
df.dropna(subset=['lunch'])
```

Replace Data Calculate the mean and replace the null value by mean by using fillna() function

```python
mean_value = df['reading score'].mean()
print(mean_value)
```

```python
df['reading score'] = df['reading score'].fillna(mean_value)
df
```
